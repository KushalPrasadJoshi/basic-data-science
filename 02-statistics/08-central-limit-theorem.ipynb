```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop_data = [np.random.randint(10, 100) for i in range(10000)] # Create a list of 10000 integers generated randomly
pop_table = pd.DataFrame({"Pop_data" : pop_data}) # Create a DataFrame with a column named Pop_data with values pop_data
```

```python
np.mean(pop_data) # Find mean of pop_data
```

Visual Representation

```python
plt.figure(figsize=(5, 5))
sns.kdeplot(x="Pop_data", data=pop_table)
plt.show()
```

Getting the mean from the sample

```python
sample_mean = []
for no_sample in range(60): # Select 60 samples
    sample_data = []
    for data in range(500): # Select 500 data in each sample
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))

np.mean(sample_mean)
    
```

```python
sample_M = pd.DataFrame({"Sample_Mean" : sample_mean})
```

Creating visualization

```python
plt.figure(figsize=(5, 5))
sns.kdeplot(x="Sample_Mean", data=sample_M)
plt.show()
```
