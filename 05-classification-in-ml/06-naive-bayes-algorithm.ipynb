```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("../assets/csv-files/suscription.csv")
dataset.head(3)
```

```python
# Check for null values
dataset.isnull().sum() # No null values
```

```python
# Visualize data
plt.figure(figsize=(5, 5))
sns.scatterplot(x="Age", y="Salary", data=dataset, hue="Purchase")
plt.show()
```

```python
# Seperate dependent and independent variables
x = dataset.iloc[:, :-1]
y = dataset["Purchase"]
```

```python
# Train test split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

#### Gaussian Naive Bayes

```python
# Build Model
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
# Visualize the decision region
plt.figure(figsize=(5, 5))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=gnb)
plt.show()
```

```python
gnb.score(x_test, y_test)
```

#### Multinomial Naive Bayes

```python
# Build Model
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB() 
mnb.fit(x_train, y_train)
```

```python
mnb.score(x_test, y_test)
```

```python
# Visualize the decision region
plt.figure(figsize=(5, 5))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=mnb)
plt.show()
```

#### Bernoulli Naive Bayes

```python
# Build Model
from sklearn.naive_bayes import BernoulliNB
```

```python
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
bnb.score(x_test, y_test)
```

```python
# Visualize the decision region
plt.figure(figsize=(5, 5))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=bnb)
plt.show() 
```
