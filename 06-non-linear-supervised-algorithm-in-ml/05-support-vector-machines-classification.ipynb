```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("../assets/csv-files/suscription.csv")
dataset.head(3)
```

```python
# Check for null values
dataset.isnull().sum() # No null values
```

```python
# Visualize the data
plt.figure(figsize=(4, 3))
sns.scatterplot(x="Age", y="Salary", data=dataset, hue="Purchase")
plt.show()
```

```python
# Seperate input and output
x = dataset.iloc[:, :-1]
y = dataset["Purchase"]
```

```python
# Train test split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=84)
```

```python
# Build the SVM model (Classification)
from sklearn.svm import SVC
```

```python
svc = SVC(kernel="linear")
svc.fit(x_train, y_train)
```

```python
svc.score(x_test, y_test)
```

```python
# Check if the data is overfitted
svc.score(x_train, y_train), svc.score(x_test, y_test)
```

```python
# Visualize decision region
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=svc)
plt.show()
```
