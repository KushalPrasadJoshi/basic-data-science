# Multi-Linear Regression

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "Age": [i for i in range(25, 50)],
    "Experience": [i for i in range(25)],
    "Salary": [35000+i*1000 for i in range(25)],
}
```

```python
dataset = pd.DataFrame(data)
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum() # Check and fill null values first
```

```python
# Visualization of data
sns.pairplot(data=dataset)
plt.show()

'''The graphs are highly linear'''
```

```python
sns.heatmap(data=dataset.corr(), annot=True)
plt.show()

'''High correlation is found so, Highly linear'''
```

```python
# Select input and output
x = dataset.iloc[:, :-1]
y = dataset["Salary"]
```

```python
# Train test split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Build the model
from sklearn.linear_model import LinearRegression
```

```python
x.ndim # Two dimensional
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train) # y = m1x1 + m2x2 
```

```python
# Test the model
lr.score(x_test, y_test)
```

```python
# New predictions
lr.predict([[50, 25]]) # [[Age, Experience]]
```

```python
lr.coef_, lr.intercept_ # y = m1x1 + m2x2
# You will get values of m1, m2 and c
# By using these values you can also do manual calculations
```
